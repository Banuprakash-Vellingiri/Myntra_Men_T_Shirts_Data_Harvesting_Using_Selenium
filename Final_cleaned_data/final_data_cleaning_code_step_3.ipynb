```python
#original_price	discounted_price	discount_percentage	total_ratings	
product_df["original_price"]=product_df["original_price"].str.replace("₹","")
product_df["discounted_price"]=product_df["discounted_price"].str.replace("₹","")
product_df["discounted_price"]=product_df["discounted_price"].str.replace("Rs. ","")
```

```python
pd.set_option('display.max_rows', None)
product_df.head(3)
```

```python
#filling NaN values for original_price and discounted_price columns
product_df.fillna({"original_price":0,"discounted_price":0},inplace=True)
#Converting original_price and discounted_price columns into numeric data type
product_df['original_price'] = pd.to_numeric(product_df['original_price'], errors='coerce')
product_df['discounted_price'] = pd.to_numeric(product_df['discounted_price'], errors='coerce')
```

```python
#Converting the "discount_percentage" column to "int" data type 
product_df["discount_percentage"]=product_df["discount_percentage"].apply(lambda x : 0).astype("int")
import math
product_df["discount_percentage"] = product_df.apply(lambda x: math.floor(((x['original_price'] - x['discounted_price']) / x['original_price']) * 100 )if x['original_price'] != 0 and x['discounted_price']!= 0  else "Nil", axis=1)
```

```python
product_df.head(2)
```

```python
#Converting Total Ratings to Numbers
def convert_to_numbers(x):
    if isinstance(x, str) and x[-1] == "k":
       converted_value=float(x[:-1])*1000
       return str(converted_value)
    else: 
        return x
product_df["total_ratings"]=product_df["total_ratings"].apply(lambda x: convert_to_numbers(x))
product_df["total_ratings"]=pd.to_numeric(product_df["total_ratings"])
# product_df["total_ratings"]=product_df["total_ratings"].astype("int64")
```

```python
#Cleaning the "available_size" column
def size(x):
    import re
    size_list=["XS","S","M","L","XXL","XL"]
    return_list=[]
    x=str(x)
    for size in size_list:
        pattern=re.search(rf"\b{size}\b",x)
        if pattern:
           return_list.append(pattern[0])  
    return (",".join(map(str,return_list )))
product_df["available_sizes"]=product_df["available_sizes"].apply(lambda x :size(x))
```

```python
#Checking NaN Values:
print(product_df.isna().sum())
```

```python
#filling NaN values
product_df.fillna("not_available",inplace=True)
print("NaN values:")
print(product_df.isna().sum())
```

```python
def removal(x):
    x=str(x)
    if x[-2:]==".0":
        return x[:-2]
    else:
        return x
product_df["total_ratings"]=product_df["total_ratings"].apply(lambda x :removal(x))
product_df["total_text_reviews"]=product_df["total_text_reviews"].apply(lambda x :removal(x))
```

```python
product_df.head(3)
```

```python
#Renaming column
product_df.rename(columns={
    "original_price": "original_price(₹)",
    "discounted_price": "discounted_price(₹)",
    "discount_percentage": "discount_percentage(%)"
}, inplace=True)
```

```python
#converting int ".csv" file
product_df.to_csv("cleaned_t_shirt_data.csv",index=False)
```
